```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Create a more original dataset
data = {
    'Year': [2015, 2016, 2017, 2018, 2019, 2020, 2021],
    'Sales': [150, 200, 250, 300, 350, 400, 450],  # Sales in thousands
    'Profit': [30, 50, 70, 90, 110, 130, 150],     # Profit in thousands
    'Category': ['Electronics', 'Electronics', 'Furniture', 'Furniture', 'Clothing', 'Clothing', 'Electronics']
}

df = pd.DataFrame(data)

# Set the style for seaborn
sns.set(style="whitegrid")

# Create a figure with 3 subplots
fig, axs = plt.subplots(3, 1, figsize=(10, 15))

# 1. Line Plot
sns.lineplot(x='Year', y='Sales', data=df, marker='o', ax=axs[0], color='blue')
axs[0].set_title('Sales Over Years')
axs[0].set_xlabel('Year')
axs[0].set_ylabel('Sales (in thousands)')

# 2. Bar Plot
sns.barplot(x='Year', y='Profit', data=df, ax=axs[1], palette='viridis')
axs[1].set_title('Profit Over Years')
axs[1].set_xlabel('Year')
axs[1].set_ylabel('Profit (in thousands)')

# 3. Scatter Plot
sns.scatterplot(x='Sales', y='Profit', hue='Category', data=df, ax=axs[2], style='Category', s=100)
axs[2].set_title('Sales vs Profit by Category')
axs[2].set_xlabel('Sales (in thousands)')
axs[2].set_ylabel('Profit (in thousands)')

# Adjust layout
plt.tight_layout()
plt.show()
```
